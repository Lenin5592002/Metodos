# Serie de Taylor y Polinomios de Lagrange CONJUNTO DE EJERCICIOS

$$ \mathbf{1.\ Función}\ f(x)=\frac{1}{25x^{2}+1},\ \text{alrededor de }\ x_{0}=0 $$


Utilizamos:
$$ P_{n}(x)=f(x_{0})+\frac{f'(x_{0})}{1!}(x-x_{0})+\frac{f''(x_{0})}{2!}(x-x_{0})^{2}+\cdots+\frac{f^{(n)}(x_{0})}{n!}(x-x_{0})^{n} $$


Como $x_0 = 0$, esto se simplifica a

$$P_{n}(x) = f(0) + \frac{f'(0)}{1!}x + \frac{f''(0)}{2!}x^{2} + \dots + \frac{f^{(n)}(0)}{n!}x^{n}.$$

Para x=0
$$f(x) = \frac{1}{25x^2 + 1}$$

Entonces:

$$f(0) = \frac{1}{25 \cdot 0^2 + 1} = \frac{1}{1} = 1.$$
entonces la constante es 1 ahora derivamos para conseguir la primera incognita:
$$f'(x) = -1 (25x^2 + 1)^{-2} \cdot (50x) = -\frac{50x}{(25x^2 + 1)^2}$$
evaluamos en x=0 entonces nos da un resultado de cero por multiplicarce numerador por cero.


Vamos a sacar la segunda derivada entonces:
$$f'(x) = -50x(25x^2 + 1)^{-2}.$$

Usamos regla del producto:

* $$u(x) = -50x, \quad u'(x) = -50$$
* $$v(x) = (25x^2 + 1)^{-2},$$
  $$v'(x) = -2(25x^2 + 1)^{-3} \cdot (50x) = -100x(25x^2 + 1)^{-3}$$

Entonces

$$\begin{aligned}f''(x) &= u'(x)v(x) + u(x)v'(x) \\&= -50(25x^2 + 1)^{-2} + (-50x)[-100x(25x^2 + 1)^{-3}] \\&= -50(25x^2 + 1)^{-2} + 5000x^2(25x^2 + 1)^{-3}.\end{aligned}$$
Evaluamos en x = 0 y nos da que es igual a -50

El término de grado 2 en el polinomio es

$$\frac{f''(0)}{2!}x^2 = \frac{-50}{2}x^2 = -25x^2.$$

Hasta aquí:

$$P_2(x) = 1 - 25x^2.$$

la tercera derivada:
$$f^{(3)}(x) = 15000x(25x^2 + 1)^{-3} - 750000x^3(25x^2 + 1)^{-4}$$
evaluada en x = 0 

la cuarta derivada:
$$f^{(4)}(x) = \frac{15000}{(25x^2 + 1)^3} - \frac{4500000x^2}{(25x^2 + 1)^4} + \frac{150000000x^4}{(25x^2 + 1)^5}$$
Y en Taylor:

$$\frac{f^{(4)}(0)}{4!}x^4 = \frac{15000}{24}x^4 = 625x^4.$$
entonces $$f(x) = 1+25x^2+625x^4.$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return 1 / (25 * x**2 + 1)

# Intervalo de graficación
x0 = 0
x = np.linspace(-2, 2, 500)   # Puedes ampliar o reducir el rango

# Evaluación de la función
y = f(x)

# Gráfica
plt.figure(figsize=(7, 5))
plt.plot(x, y, label=r"$f(x)=\frac{1}{25x^2+1}$")
plt.title("Gráfica de la función 1 / (25x^2 + 1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función original
def f(x):
    return 1 / (25 * x**2 + 1)

# Aproximación de Taylor orden 4
def T4(x):
    return 1 - 25*x**2 + (25**2)*(x**4)

# Aproximación de Taylor orden 2
def T2(x):
    return 1 - 25*x**2

# Rango de graficación
x = np.linspace(-1, 1, 500)   # Intervalo centrado en x0 = 0

# Evaluaciones
y_f = f(x)
y_T4 = T4(x)
y_T2 = T2(x)

# Graficar
plt.figure(figsize=(7, 5))
plt.plot(x, y_f, label=r"$f(x)=\frac{1}{25x^2+1}$", linewidth=2)
plt.plot(x, y_T4, "--", label=r"$T_4(x)=1-25x^2+25^2 x^4$")
plt.plot(x, y_T2, "--", label=r"$T_2(x)=1-25x^2$")

plt.title("Función y aproximaciones de Taylor")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

Ahora para polinomio de lagrange:
Como el polinomio de Lagrange **necesita puntos** $(x_i, f(x_i))$, vamos a elegir 3 puntos alrededor de 0 para hacer un polinomio **cuadrático** (grado 2):

$$x_0 = -\frac{1}{5}, \quad x_1 = 0, \quad x_2 = \frac{1}{5}.$$

entoces
$$f(x) = \frac{1}{25x^2 + 1}$$

* En $x_0 = -\frac{1}{5}$:

$$f(x_0) = f\left(-\frac{1}{5}\right) = \frac{1}{25 \left(\frac{1}{25}\right) + 1} = \frac{1}{1+1} = \frac{1}{2}$$

* En $x_1 = 0$:

$$f(x_1) = f(0) = \frac{1}{1} = 1$$

* En $x_2 = \frac{1}{5}$:

$$f(x_2) = f\left(\frac{1}{5}\right) = \frac{1}{25 \left(\frac{1}{25}\right) + 1} = \frac{1}{2} = \frac{1}{2}$$

Entonces los puntos son:

$$(x_0, y_0) = \left(-\frac{1}{5}, \frac{1}{2}\right), \quad (x_1, y_1) = (0, 1), \quad (x_2, y_2) = \left(\frac{1}{5}, \frac{1}{2}\right).$$

$$P_2(x) = y_0L_0(x) + y_1L_1(x) + y_2L_2(x),$$

donde

$$L_k(x) = \prod_{\substack{j=0 \\ j\neq k}}^{2} \frac{x - x_j}{x_k - x_j}.$$

Los calculamos uno por uno.

**$L_0(x)$**

$$
\begin{aligned}
L_0(x) &= \frac{x - x_1}{x_0 - x_1} \cdot \frac{x - x_2}{x_0 - x_2} = \frac{x - 0}{-\frac{1}{5} - 0} \cdot \frac{x - \frac{1}{5}}{-\frac{1}{5} - \frac{1}{5}} \\
L_0(x) &= \frac{x}{-\frac{1}{5}} \cdot \frac{x - \frac{1}{5}}{-\frac{2}{5}} = (-5x) \cdot \left(-\frac{5}{2}\right) \left(x - \frac{1}{5}\right) = \frac{25}{2}x(x - \frac{1}{5})
\end{aligned}
$$

$$
{L_0(x) = \frac{5x(5x - 1)}{2}}
$$

**$L_1(x)$**

$$
\begin{aligned}
L_1(x) &= \frac{x - x_0}{x_1 - x_0} \cdot \frac{x - x_2}{x_1 - x_2} = \frac{x + \frac{1}{5}}{0 + \frac{1}{5}} \cdot \frac{x - \frac{1}{5}}{0 - \frac{1}{5}} \\
L_1(x) &= 5(x + \frac{1}{5}) \cdot [-5(x - \frac{1}{5})] = -25(x + \frac{1}{5})(x - \frac{1}{5}) = -25(x^2 - \frac{1}{25})
\end{aligned}
$$

$$
{L_1(x) = 1 - 25x^2}
$$

**$L_2(x)$**

$$
\begin{aligned}
L_2(x) &= \frac{x - x_0}{x_2 - x_0} \cdot \frac{x - x_1}{x_2 - x_1} = \frac{x + \frac{1}{5}}{\frac{1}{5} + \frac{1}{5}} \cdot \frac{x - 0}{\frac{1}{5} - 0} \\
L_2(x) &= \frac{x + \frac{1}{5}}{\frac{2}{5}} \cdot 5x = \frac{5}{2}(x + \frac{1}{5}) \cdot 5x = \frac{25}{2}x(x + \frac{1}{5})
\end{aligned}
$$

$$
{L_2(x) = \frac{5x(5x + 1)}{2}}
$$

Sustituimos:

$$P_2(x) = \frac{1}{2} \cdot \frac{5x(5x - 1)}{2} + (1 - 25x^2) + \frac{1}{2} \cdot \frac{5x(5x + 1)}{2}$$

simplificamos:

$$
{P_2(x) = 1 - \frac{25}{2}x^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Función original
def f(x):
    return 1 / (25 * x**2 + 1)

# Polinomio P2
def P2(x):
    return 1 - (25/2) * x**2

# Rango de graficación
x = np.linspace(-1, 1, 500)

# Evaluaciones
y_f = f(x)
y_P2 = P2(x)

# Graficar
plt.figure(figsize=(7, 5))
plt.plot(x, y_f, label=r"$f(x)=\frac{1}{25x^2+1}$", linewidth=2)
plt.plot(x, y_P2, "--", label=r"$P_2(x)=1-\frac{25}{2}x^2$", linewidth=2)

plt.title("Función original y polinomio $P_2(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

### 2.derivadas

$$f'(x) = \frac{1}{1 + x^2}.$$

$$f''(x) = -1 (1 + x^2)^{-2} \cdot (2x) = -\frac{2x}{(1 + x^2)^2}.$$

$$
\begin{aligned}
f^{(3)}(x) = -2(1 + x^2)^{-2} + 8x^2(1 + x^2)^{-3}.
\end{aligned}
$$


$$P_3(x) = \frac{\pi}{4} + \frac{1}{2}(x-1) - \frac{1}{4}(x-1)^2 + \frac{1}{12}(x-1)^3$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Función arctan(x)
def f(x):
    return np.arctan(x)

x0 = 1

# Rango de graficación
x = np.linspace(x0 - 4, x0 + 4, 500)
y = f(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, label=r"$f(x)=\arctan(x)$", linewidth=2)

# Punto x0
plt.scatter([x0], [f(x0)], label=r"$x_0=1$")

plt.title(r"Gráfica de $f(x)=\arctan(x)$ alrededor de $x_0=1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

plt.show()          # 👈 MUY importante
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Función original
def f(x):
    return np.arctan(x)

# Polinomio de Taylor P3 alrededor de x0=1
def P3(x):
    return (
        math.pi/4 
        + 0.5*(x - 1)
        - 0.25*(x - 1)**2
        + (1/12)*(x - 1)**3
    )

x0 = 1

# Rango de graficación
x = np.linspace(x0 - 4, x0 + 4, 500)

# Evaluaciones
y_f = f(x)
y_P3 = P3(x)

# Gráfica
plt.figure(figsize=(7, 5))
plt.plot(x, y_f, label=r"$f(x)=\arctan(x)$", linewidth=2)
plt.plot(x, y_P3, "--", label=r"$P_3(x)$", linewidth=2)

plt.scatter([x0], [f(x0)], color="red", label=r"$x_0=1$")

plt.title(r"Función $f(x)=\arctan(x)$ y polinomio $P_3(x)$ en $x_0=1$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
