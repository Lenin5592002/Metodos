### 1.	Calcule los errores absoluto y relativo en las aproximaciones de p por p*
### a.	p = œÄ, p* = 22/ 7 
### b.  p = œÄ, p* = 3.1416
### c.  p = e, p* = 2.718
### d.  p = $\sqrt{2}$, p* = 1.1414
$$ E_a = |\, x_{\text{medido}} - x_{\text{verdadero}} \,| $$
$$E_r = \frac{E_a}{|\, x_{\text{verdadero}} \,|} = \frac{|\, x_{\text{medido}} - x_{\text{verdadero}} \,|}{|\, x_{\text{verdadero}} \,|}$$



```python
import math


def calcular_errores(p, p_aprox):
    error_absoluto = abs(p - p_aprox)
    error_relativo = error_absoluto / abs(p)
    return error_absoluto, error_relativo


casos = {
    "a": (math.pi, 22/7),
    "b": (math.pi, 3.1416),
    "c": (math.e, 2.718),
    "d": (math.sqrt(2), 1.414)
}


print("C√°lculo de errores absoluto y relativo\n")
for letra, (p, p_aprox) in casos.items():
    ea, er = calcular_errores(p, p_aprox)
    print(f"({letra}) p = {p:.10f}, p* = {p_aprox}")
    print(f"    Error absoluto = {ea:.10e}")
    print(f"    Error relativo = {er:.10e}\n")
```

### 3. Encuentre el intervalo m√°s largo en el que se debe encontrar ùëù‚àó para aproximarse a ùëù con error relativo m√°ximo de 10‚àí4 para cada valor de ùëù.
### a. $ œÄ $
### b. $ e $
### c. $\sqrt{2}$
### $\text{d. } \sqrt[3]{7}$
$$
|p - p^{*}| \leq 10^{-4} \cdot |p|
$$



```python
import math

error_relativo_max = 1e-4

valores = {
    "a": ("œÄ", math.pi),
    "b": ("e", math.e),
    "c": ("‚àö2", math.sqrt(2)),
    "d": ("‚àõ7", 7 ** (1/3))
}

print("=== EJERCICIO 3 ===")
print("Intervalo m√°s largo para p* con error relativo m√°ximo de 1e-4\n")

for letra, (nombre, p) in valores.items():
  
    delta = error_relativo_max * abs(p)
    p_min = p - delta
    p_max = p + delta
    longitud = p_max - p_min

    print(f"({letra}) p = {nombre} = {p:.10f}")
    print(f"    Intervalo: [{p_min:.10f}, {p_max:.10f}]")
    print(f"    Longitud del intervalo: {longitud:.10e}\n")
```

### 4. Use la aritm√©tica de redondeo de tres d√≠gitos para realizar lo siguiente. Calcule los errores absoluto y relativo con el valor exacto determinado para por lo menos cinco d√≠gitos. 
$$

$$

$$

$$
\text{d. } \frac{\sqrt{13} + \sqrt{11}}{\sqrt{13} - \sqrt{11}}
$$


```python
import math


def redondear3(x):
    if x == 0:
        return 0.0
    else:
        return round(x, -int(math.floor(math.log10(abs(x)))) + 2)


def errores(p_real, p_aprox):
    ea = abs(p_real - p_aprox)
    er = ea / abs(p_real)
    return ea, er


print("=== EJERCICIO 4 ===\n")


num_a = redondear3(13/14)
resta_a = redondear3(num_a - redondear3(5/7))
den_a = redondear3(2 * redondear3(math.e))
den_a = redondear3(den_a - 5.4)
aprox_a = redondear3(resta_a / den_a)
exact_a = (13/14 - 5/7) / (2*math.e - 5.4)
ea_a, er_a = errores(exact_a, aprox_a)

print(f"a) Aprox. = {aprox_a:.5f}, Exacto = {exact_a:.5f}")
print(f"   Error absoluto = {ea_a:.5e}")
print(f"   Error relativo = {er_a:.5e}\n")



term1_b = redondear3(-10 * redondear3(math.pi))
term2_b = redondear3(6 * redondear3(math.e))
term3_b = redondear3(-3/61)
aprox_b = redondear3(term1_b + redondear3(term2_b + term3_b))
exact_b = -10*math.pi + 6*math.e - 3/61
ea_b, er_b = errores(exact_b, aprox_b)

print(f"b) Aprox. = {aprox_b:.5f}, Exacto = {exact_b:.5f}")
print(f"   Error absoluto = {ea_b:.5e}")
print(f"   Error relativo = {er_b:.5e}\n")


term1_c = redondear3(2/9)
term2_c = redondear3(9/11)
aprox_c = redondear3(term1_c * term2_c)
exact_c = (2/9) * (9/11)
ea_c, er_c = errores(exact_c, aprox_c)

print(f"c) Aprox. = {aprox_c:.5f}, Exacto = {exact_c:.5f}")
print(f"   Error absoluto = {ea_c:.5e}")
print(f"   Error relativo = {er_c:.5e}\n")



num_d = redondear3(redondear3(math.sqrt(13)) + redondear3(math.sqrt(11)))
den_d = redondear3(redondear3(math.sqrt(13)) - redondear3(math.sqrt(11)))
aprox_d = redondear3(num_d / den_d)
exact_d = (math.sqrt(13) + math.sqrt(11)) / (math.sqrt(13) - math.sqrt(11))
ea_d, er_d = errores(exact_d, aprox_d)

print(f"d) Aprox. = {aprox_d:.5f}, Exacto = {exact_d:.5f}")
print(f"   Error absoluto = {ea_d:.5e}")
print(f"   Error relativo = {er_d:.5e}\n")
```

### **5.** Los primeros tres t√©rminos diferentes a cero de la serie de Maclaurin para la funci√≥n arcotangente son:  

$$
x - \frac{1}{3}x^{3} + \frac{1}{5}x^{5}
$$  

### Calcule los **errores absoluto y relativo** en las siguientes aproximaciones de $ œÄ $ mediante el polinomio en lugar del arcotangente:

$$
\text{a. } 4\left[\arctan\left(\frac{1}{2}\right) + \arctan\left(\frac{1}{3}\right)\right]
$$  

$$
\text{b. } 16\arctan\left(\frac{1}{5}\right) - 4\arctan\left(\frac{1}{239}\right)
$$


```python

import math

def arctan_poly_3(x: float) -> float:
    """Aproxima arctan(x) con los 3 primeros t√©rminos: x - x^3/3 + x^5/5."""
    return x - (x**3)/3 + (x**5)/5

def errores(val_real: float, val_aprox: float):
    ea = abs(val_real - val_aprox)
    er = ea / abs(val_real)
    return ea, er

PI = math.pi


pi_aprox_a = 4 * (arctan_poly_3(1/2) + arctan_poly_3(1/3))
ea_a, er_a = errores(PI, pi_aprox_a)


pi_aprox_b = 16 * arctan_poly_3(1/5) - 4 * arctan_poly_3(1/239)
ea_b, er_b = errores(PI, pi_aprox_b)

print("=== Aproximaciones de œÄ usando arctan(x) ‚âà x - x^3/3 + x^5/5 ===\n")

print("(a) 4[arctan(1/2) + arctan(1/3)]")
print(f"    œÄ_aprox = {pi_aprox_a:.15f}")
print(f"    Error absoluto = {ea_a:.3e}")
print(f"    Error relativo = {er_a:.3e}\n")

print("(b) 16 arctan(1/5) - 4 arctan(1/239)")
print(f"    œÄ_aprox = {pi_aprox_b:.15f}")
print(f"    Error absoluto = {ea_b:.3e}")
print(f"    Error relativo = {er_b:.3e}")
```

**6.** El n√∫mero \( e \) se puede definir por medio de  

$$
e = \sum_{n=0}^{\infty} \frac{1}{n!},
$$  

donde  n! = n(n - 1)....*2*1 para $
n \neq 0
$
  y 0! = 1 .  
Calcule los **errores absoluto y relativo** en la siguiente aproximaci√≥n de \( e \):

$$
\text{a. } \sum_{n=0}^{5} \frac{1}{n!}
$$  

$$
\text{b. } \sum_{n=0}^{10} \frac{1}{n!}
$$


```python
import math

def aproximar_e(n):
    """Calcula la suma de 1/n! desde n=0 hasta el valor dado."""
    suma = 0
    for k in range(n + 1):
        suma += 1 / math.factorial(k)
    return suma

def errores(p_real, p_aprox):
    ea = abs(p_real - p_aprox)
    er = ea / abs(p_real)
    return ea, er


e_real = math.e


e_aprox_5 = aproximar_e(5)
ea_5, er_5 = errores(e_real, e_aprox_5)


e_aprox_10 = aproximar_e(10)
ea_10, er_10 = errores(e_real, e_aprox_10)

print("=== Aproximaci√≥n del n√∫mero e mediante sumatoria 1/n! ===\n")

print("(a) Suma hasta n = 5")
print(f"    e_aprox = {e_aprox_5:.10f}")
print(f"    e_real  = {e_real:.10f}")
print(f"    Error absoluto = {ea_5:.5e}")
print(f"    Error relativo = {er_5:.5e}\n")

print("(b) Suma hasta n = 10")
print(f"    e_aprox = {e_aprox_10:.10f}")
print(f"    e_real  = {e_real:.10f}")
print(f"    Error absoluto = {ea_10:.5e}")
print(f"    Error relativo = {er_10:.5e}\n")
```

### **7.** Suponga que dos puntos $(x_0, y_0)$ y $(x_1, y_1)$ se encuentran en l√≠nea recta con $y_1 \neq y_0$.  Existen dos f√≥rmulas para encontrar la intersecci√≥n $x$ de la l√≠nea:
$$
x = \frac{x_0 y_1 - x_1 y_0}{y_1 - y_0}
\quad \text{y} \quad
x = x_0 - \frac{(x_1 - x_0) y_0}{y_1 - y_0}
$$
### **a.** Use los datos  $ (x_0, y_0) = (1.31, 3.24) $ y  $ (x_1, y_1) = (1.93, 5.76) $ ,  y la aritm√©tica de redondeo de tres d√≠gitos para calcular la intersecci√≥n con \( x \) de ambas maneras.  ¬øCu√°l m√©todo es mejor y por qu√©?


```python

import math


def round_sig3(x: float) -> float:
    """Redondea a 3 cifras significativas (simula aritm√©tica de 3 d√≠gitos)."""
    if x == 0:
        return 0.0
    return round(x, -int(math.floor(math.log10(abs(x)))) + 2)

def interseccion_forma1_r3(x0, y0, x1, y1):

    a = round_sig3(x0 * y1)
    b = round_sig3(x1 * y0)
    num = round_sig3(a - b)
    den = round_sig3(y1 - y0)
    return round_sig3(num / den)

def interseccion_forma2_r3(x0, y0, x1, y1):
    dx = round_sig3(x1 - x0)
    prod = round_sig3(dx * y0)
    den = round_sig3(y1 - y0)
    frac = round_sig3(prod / den)
    return round_sig3(x0 - frac)

def interseccion_exacta(x0, y0, x1, y1):

    return (x0*y1 - x1*y0) / (y1 - y0)

def errores(real, aprox):
    ea = abs(real - aprox)
    er = ea / abs(real) if real != 0 else float('inf')
    return ea, er

x0, y0 = 1.31, 3.24
x1, y1 = 1.93, 5.76


x_f1 = interseccion_forma1_r3(x0, y0, x1, y1)
x_f2 = interseccion_forma2_r3(x0, y0, x1, y1)
x_real = interseccion_exacta(x0, y0, x1, y1)

ea1, er1 = errores(x_real, x_f1)
ea2, er2 = errores(x_real, x_f2)

print("=== Intersecci√≥n en x con aritm√©tica de 3 cifras ===\n")
print(f"Datos: (x0,y0)=( {x0}, {y0} ), (x1,y1)=( {x1}, {y1} )")
print(f"Valor exacto (doble precisi√≥n): x = {x_real:.10f}\n")

print("F√≥rmula 1: x = (x0*y1 - x1*y0)/(y1 - y0)")
print(f"  x ‚âà {x_f1:.10f}")
print(f"  Error absoluto = {ea1:.5e}")
print(f"  Error relativo = {er1:.5e}\n")

print("F√≥rmula 2: x = x0 - ((x1 - x0)*y0)/(y1 - y0)")
print(f"  x ‚âà {x_f2:.10f}")
print(f"  Error absoluto = {ea2:.5e}")
print(f"  Error relativo = {er2:.5e}\n")

mejor = "F√≥rmula 1" if ea1 < ea2 else "F√≥rmula 2"
print(f"‚áí Mejor con 3 d√≠gitos: {mejor}")
print("Motivo: la que evita mayor cancelaci√≥n en el numerador produce menos error de redondeo.")
```
