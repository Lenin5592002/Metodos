### Caso $$cos(x), x_0=0$$

- $$f(x)=cos(x)$$
- $$f'(x)=-sen(x)$$
- $$f''(x)=-cos(x)$$
- $$f'''(x)=sen(x)$$
- $$f^{(4)}(x)=cos(x)$$




- $$f(x) \approx \frac{1}{24}(x)^4 - \frac{1}{2}(x)^2 + 1 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Calcular cos(x)
y = np.cos(x)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='cos(x)', linewidth=2)

# Agregar detalles
plt.title('Gráfica de cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el rango de x
x = np.linspace(-3, 3, 400)

# Definimos las funciones
y1 = np.cos(x)
y2 = -0.5 * x**2 + 1

# Buscamos las intersecciones (donde la diferencia cambia de signo)
diff = y1 - y2
indices = np.where(np.diff(np.sign(diff)))[0]

# Calculamos los puntos aproximados de intersección
x_inter = []
y_inter = []
for i in indices:
    # Aproximamos el punto de cruce linealmente
    xi = (x[i] + x[i+1]) / 2
    yi = np.cos(xi)
    x_inter.append(xi)
    y_inter.append(yi)

# Graficamos
plt.figure(figsize=(8,5))
plt.plot(x, y1, label='cos(x)', linewidth=2)
plt.plot(x, y2, label='-(1/2)x² + 1', linewidth=2, linestyle='--')
plt.scatter(x_inter, y_inter, color='red', zorder=5, label='Intersección')

# Detalles de la gráfica
plt.title('Intersección entre cos(x) y -½x² + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()

# Mostrar coordenadas de los puntos de intersección
for xi, yi in zip(x_inter, y_inter):
    print(f"Intersección aproximada en x = {xi:.3f}, y = {yi:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(-4, 4, 600)

# Definición de funciones
y1 = np.cos(x)
y2 = -0.5 * x**2 + 1
y3 = (1/24) * x**4 - 0.5 * x**2 + 1

# Diferencia entre cos(x) y la nueva función para detectar intersecciones
diff = y1 - y3
indices = np.where(np.diff(np.sign(diff)))[0]

# Aproximar puntos de intersección
x_inter = []
y_inter = []
for i in indices:
    xi = (x[i] + x[i+1]) / 2
    yi = np.cos(xi)
    x_inter.append(xi)
    y_inter.append(yi)

# Gráfica
plt.figure(figsize=(10,6))
plt.plot(x, y1, label='cos(x)', linewidth=2.5, color='blue')
plt.plot(x, y2, label='-(1/2)x² + 1', linewidth=2.2, linestyle='--', color='green')
plt.plot(x, y3, label='(1/24)x⁴ - (1/2)x² + 1', linewidth=2.2, linestyle='-.', color='orange')

# Puntos de intersección (más notorios)
plt.scatter(x_inter, y_inter, color='red', s=120, edgecolor='black', zorder=5, label='Intersecciones')

# Personalización
plt.title('Intersección entre cos(x), -½x² + 1 y (1/24)x⁴ - ½x² + 1', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend(fontsize=10)
plt.ylim(-2, 2)
plt.show()

# Mostrar coordenadas aproximadas
for xi, yi in zip(x_inter, y_inter):
    print(f"Intersección aproximada en x = {xi:.3f}, y = {yi:.3f}")
```

## Parte 2 del taller

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:

- $$ (0, 0),\ (30, 0.5),\ (60, \frac{\sqrt{3}}{2}),\ (90, 1) $$


$$P_1(x)=
\frac{1}{2}\,\frac{x(x-60)(x-90)}{54000}
\;+\;
\frac{\sqrt{3}}{2}\,\frac{x(x-30)(x-90)}{-54000}
\;+\;
\frac{x(x-30)(x-60)}{162000}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos
x = [0, 30, 60, 90]
y = [0, 0.5, np.sqrt(3)/2, 1]

# Crear la gráfica solo con puntos
plt.figure(figsize=(7, 5))
plt.scatter(x, y, color='blue', s=80, label='Puntos dados')

# Añadir etiquetas y detalles
plt.title('Gráfico de los puntos dados')
plt.xlabel('x (grados)')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos
x_puntos = np.array([0, 30, 60, 90])
y_puntos = np.array([0, 0.5, np.sqrt(3)/2, 1])

# Definir el polinomio P1(x)
def P1(x):
    term1 = (1/2) * x * (x - 60) * (x - 90) / 54000
    term2 = (np.sqrt(3)/2) * x * (x - 30) * (x - 90) / (-54000)
    term3 = x * (x - 30) * (x - 60) / 162000
    return term1 + term2 + term3

# Generar valores de x para la curva
x = np.linspace(0, 90, 400)
y = P1(x)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.scatter(x_puntos, y_puntos, color='red', s=80, label='Puntos dados')
plt.plot(x, y, color='blue', linewidth=2, label='Polinomio P₁(x)')

# Detalles del gráfico
plt.title('Interpolación de Lagrange - P₁(x)')
plt.xlabel('x (grados)')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
