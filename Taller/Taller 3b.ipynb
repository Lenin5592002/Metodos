## **Grafico para el sistema de ecuaciones lineales:
$$x_1 + x_2 = 7$$
$$-2x_1 + 5x_2 = 0$$

* $$X_0 = (0, 0)$$

```python

if __name__ == "__main__":
    # 1. Definir sistema
    A = [[1, 1], [-2, 5]]
    b = [7, 0]
    x0 = [0, 0] 
    
    # 2. Calcular (Devuelve matrices grandes con todo el camino)
    path_jacobi = gauss_jacobi(A=A, b=b, x=x0, tol=1e-6, max_iter=10)
    path_gs = gauss_seidel(A=A, b=b, x=x0, tol=1e-6, max_iter=10)

    # 3. Configurar fondo
    x_vals = np.linspace(-1, 8, 100)
    y_eq1 = 7 - x_vals        
    y_eq2 = 0.4 * x_vals      

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # --- GRFICA 1: JACOBI ---
    ax1.plot(x_vals, y_eq1, 'b-', alpha=0.3, label='x1+x2=7')
    ax1.plot(x_vals, y_eq2, 'r-', alpha=0.3, label='-2x1+5x2=0')
    # Ahora esto funcionar porque path_jacobi tiene el tamao correcto
    ax1.plot(path_jacobi[:,0], path_jacobi[:,1], 'o-', color='purple', label='Trayectoria Jacobi')
    ax1.plot(5, 2, 'k*', markersize=15, label='Solucin Exacta')
    
    # Flechas
    for i in range(len(path_jacobi)-1):
        ax1.annotate('', xy=path_jacobi[i+1], xytext=path_jacobi[i],
                     arrowprops=dict(arrowstyle='->', color='purple', lw=1.5))
    ax1.set_title('Mtodo de Jacobi')
    ax1.grid(True); ax1.legend()

    # --- GRFICA 2: GAUSS-SEIDEL ---
    ax2.plot(x_vals, y_eq1, 'b-', alpha=0.3, label='x1+x2=7')
    ax2.plot(x_vals, y_eq2, 'r-', alpha=0.3, label='-2x1+5x2=0')
    ax2.plot(path_gs[:,0], path_gs[:,1], 'o-', color='green', label='Trayectoria G-S')
    ax2.plot(5, 2, 'k*', markersize=15, label='Solucin Exacta')
    
    # Flechas
    for i in range(len(path_gs)-1):
        ax2.annotate('', xy=path_gs[i+1], xytext=path_gs[i],
                     arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
    ax2.set_title('Mtodo de Gauss-Seidel')
    ax2.grid(True); ax2.legend()

    plt.tight_layout()
    plt.show()
```

## **Ejercicio 2**
$$
\begin{cases}
x_1 + x_2 = 6 \\
-2x_1 + x_2 = 0
\end{cases}
$$

**Forma Matricial ($Ax = b$)**

$$
A = \begin{pmatrix} 1 & 1 \\ -2 & 1 \end{pmatrix}, \quad b = \begin{pmatrix} 6 \\ 0 \end{pmatrix}
$$

**Solucin Analtica**

El punto donde las rectas se cruzan es $(2, 4)$.

**Nota importante sobre Convergencia:** La matriz $A$ **no es diagonalmente dominante**.

* En la fila 2, el valor absoluto del elemento no diagonal ($|-2| = 2$) es mayor que el elemento diagonal ($|1| = 1$).

```python
if __name__ == "__main__":

    A = [[1, 1], 
         [-2, 1]]
    b = [6, 0]
    

    x0 = [0, 0] 
    

    path_jacobi = gauss_jacobi(A=A, b=b, x=x0, tol=1e-6, max_iter=10)
    path_gs = gauss_seidel(A=A, b=b, x=x0, tol=1e-6, max_iter=10)


    x_vals = np.linspace(-2, 6, 100)
    
  
    y_eq1 = 6 - x_vals        
 
    y_eq2 = 2 * x_vals      

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))


    ax1.plot(x_vals, y_eq1, 'b-', alpha=0.3, label='x1+x2=6')
    ax1.plot(x_vals, y_eq2, 'r-', alpha=0.3, label='-2x1+x2=0')
    
    ax1.plot(path_jacobi[:,0], path_jacobi[:,1], 'o-', color='purple', label='Trayectoria Jacobi')
    ax1.plot(2, 4, 'k*', markersize=15, label='Solucin Exacta (2,4)')
    

    for i in range(len(path_jacobi)-1):
        ax1.annotate('', xy=path_jacobi[i+1], xytext=path_jacobi[i],
                     arrowprops=dict(arrowstyle='->', color='purple', lw=1.5))
    
    ax1.set_title(f'Ejercicio 2: Jacobi\nInicio: {x0}')
    ax1.grid(True); ax1.legend()


    ax2.plot(x_vals, y_eq1, 'b-', alpha=0.3, label='x1+x2=6')
    ax2.plot(x_vals, y_eq2, 'r-', alpha=0.3, label='-2x1+x2=0')
    
    ax2.plot(path_gs[:,0], path_gs[:,1], 'o-', color='green', label='Trayectoria G-S')
    ax2.plot(2, 4, 'k*', markersize=15, label='Solucin Exacta (2,4)')
    

    for i in range(len(path_gs)-1):
        ax2.annotate('', xy=path_gs[i+1], xytext=path_gs[i],
                     arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
    
    ax2.set_title(f'Ejercicio 2: Gauss-Seidel\nInicio: {x0}')
    ax2.grid(True); ax2.legend()

    plt.tight_layout()
    plt.show()
    
    print("ltimo punto Jacobi:", path_jacobi[-1])
    print("ltimo punto GS:", path_gs[-1])
```

## **Ejercicio 3**


```python
if __name__ == "__main__":
    # 1. Definir el sistema 4x4
    A = [
        [ 1.0, -1.0,  2.0,  0.0],
        [-1.0,  1.0, -1.0,  3.0],
        [ 2.0, -1.0,  1.0, -1.0],
        [ 0.0,  3.0, -1.0,  2.0]
    ]
    b = [1360.0, 1130.0, 1350.0, 3650.0]

    # 2. Las 3 pruebas iniciales
    pruebas = [
        np.array([0.0, 0.0, 0.0, 0.0]),         # Prueba 1
        np.array([100.0, 100.0, 100.0, 100.0]), # Prueba 2
        np.array([10.0, -10.0, 50.0, -50.0])    # Prueba 3
    ]

    # 3. Preparamos 3 grficos (uno para cada prueba)
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    print(f"{'PRUEBA':<10} | {'JACOBI':<12} | {'GAUSS-SEIDEL'}")
    print("-" * 45)

    # 4. Bucle para probar cada punto inicial
    for i, x0 in enumerate(pruebas):
        # Ejecutamos los mtodos (Limitamos a 15 iteraciones porque diverge rpido)
        path_jacobi = gauss_jacobi(A=A, b=b, x=x0, tol=1e-6, max_iter=15)
        path_gs = gauss_seidel(A=A, b=b, x=x0, tol=1e-6, max_iter=15)

        # Calculamos la MAGNITUD (Norma) del vector en cada paso
        # Esto nos permite graficar en 2D algo que es 4D
        normas_jacobi = [np.linalg.norm(p) for p in path_jacobi]
        normas_gs = [np.linalg.norm(p) for p in path_gs]

        # --- GRAFICAR ---
        ax = axes[i]
        ax.plot(normas_jacobi, 'o-', color='purple', label='Jacobi')
        ax.plot(normas_gs, 's-', color='green', label='Gauss-Seidel')
        
        ax.set_title(f"Prueba {i+1}: Inicio {x0}")
        ax.set_xlabel("Iteracin")
        ax.set_ylabel("Tamao del Error (Norma)")
        ax.grid(True)
        ax.legend()

        # Diagnstico simple para imprimir
        # Si el ltimo valor es mayor que el primero, est creciendo (diverge)
        estado_j = "DIVERGE " if normas_jacobi[-1] > normas_jacobi[0] else "Converge"
        estado_gs = "DIVERGE " if normas_gs[-1] > normas_gs[0] else "Converge"
        
        print(f"Prueba {i+1}   | {estado_j:<12} | {estado_gs}")

    plt.tight_layout()
    plt.show()
```
